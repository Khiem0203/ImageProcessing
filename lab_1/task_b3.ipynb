```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_original = cv2.imread("./cat.jpg")
gray_img = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)
cv2.imwrite("./output_images_b3/cat_gray.jpg", gray_img)
#Plot grayscale image
plt.title("Grayscale Image")
plt.axis("off")
plt.imshow(gray_img, cmap="gray")
```

```python
hitsto_gray = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
plt.title("Histogram Grayscale Image")
plt.plot(hitsto_gray)
plt.show()
```

```python
#Define 3 thresholds
T1 = 80
T2 = 120
T3 = 160

# Apply thresholding
# 255: max value of a pixel
# > threshold = 255 = white; < threshold  = 0 = black
_, binary1 = cv2.threshold(gray_img, T1, 255, cv2.THRESH_BINARY)
_, binary2 = cv2.threshold(gray_img, T2, 255, cv2.THRESH_BINARY)
_, binary3 = cv2.threshold(gray_img, T3, 255, cv2.THRESH_BINARY)
```

```python
plt.figure(figsize=(16, 4))

plt.subplot(1, 4, 1)
plt.imshow(gray_img, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

plt.subplot(1, 4, 2)
plt.imshow(binary1, cmap="gray")
plt.title(f"Threshold = {T1}")
cv2.imwrite("./output_images_b3/T80.jpg", binary1)
plt.axis("off")

plt.subplot(1, 4, 3)
plt.imshow(binary2, cmap="gray")
plt.title(f"Threshold = {T2}")
cv2.imwrite("./output_images_b3/T120.jpg", binary2)
plt.axis("off")

plt.subplot(1, 4, 4)
plt.imshow(binary3, cmap="gray")
plt.title(f"Threshold = {T3}")
cv2.imwrite("./output_images_b3/T160.jpg", binary3)
plt.axis("off")

plt.tight_layout()
plt.show()
```
