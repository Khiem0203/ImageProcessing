```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_original = cv2.imread("./cat.jpg")
og_img = cv2.cvtColor(img_original, cv2.COLOR_BGR2RGB)
#Plot origin image
plt.title("Original Image")
plt.axis("off")
plt.imshow(og_img)
```

```python
#Convert to grayscale
gray_img = cv2.cvtColor(img_original, cv2.COLOR_BGR2GRAY)
cv2.imwrite("./output_images_b2/cat_gray.jpg", gray_img)
#Plot grayscale image
plt.title("Grayscale Image")
plt.axis("off")
plt.imshow(gray_img, cmap="gray")
```

```python
def adjust_brightness_contrast(img, a, b):
    img_processed = cv2.convertScaleAbs(img, alpha=a, beta=b)
    return img_processed
```

```python
params = [
    (1.0,  50),  
    (1.0, -50), 
    (1.5,   0),
    (0.5,   0), 
]

img1 = adjust_brightness_contrast(gray_img, 1.0, 50)  
img2 = adjust_brightness_contrast(gray_img, 1.0, -50)  
img3 = adjust_brightness_contrast(gray_img, 1.5, 0)    
img4 = adjust_brightness_contrast(gray_img, 0.5, 0)  
```

```python
plt.figure(figsize=(18, 4))

plt.subplot(1, 5, 1)
plt.imshow(gray_img, cmap="gray")
plt.title("Original")
plt.axis("off")

plt.subplot(1, 5, 2)
plt.imshow(img1, cmap="gray")
plt.title("(a,b) = (1.0, +50)")
cv2.imwrite("./output_images_b2/1_50.jpg", img1)
plt.axis("off")

plt.subplot(1, 5, 3)
plt.imshow(img2, cmap="gray")
plt.title("(a,b) = (1.0, -50)")
cv2.imwrite("./output_images_b2/1_minus50.jpg", img2)
plt.axis("off")

plt.subplot(1, 5, 4)
plt.imshow(img3, cmap="gray")
plt.title("(a,b) = (1.5, 0)")
cv2.imwrite("./output_images_b2/1point5_0.jpg", img3)
plt.axis("off")

plt.subplot(1, 5, 5)
plt.imshow(img4, cmap="gray")
plt.title("(a,b) = (0.5, 0)")
cv2.imwrite("./output_images_b2/0point5_0.jpg", img3)
plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(18, 3))

plt.subplot(1, 5, 1)
plt.hist(gray_img.ravel(), bins=256, range=(0, 255), color="blue", alpha=0.7)
plt.title("Original")
plt.xlim(0, 255)

plt.subplot(1, 5, 2)
plt.hist(img1.ravel(), bins=256, range=(0, 255), color="green", alpha=0.7)
plt.title("(a,b) = (1.0, +50)")
plt.xlim(0, 255)

plt.subplot(1, 5, 3)
plt.hist(img2.ravel(), bins=256, range=(0, 255), color="red", alpha=0.7)
plt.title("(a,b) = (1.0, -50)")
plt.xlim(0, 255)

plt.subplot(1, 5, 4)
plt.hist(img3.ravel(), bins=256, range=(0, 255), color="orange", alpha=0.7)
plt.title("(a,b) = (1.5, 0)")
plt.xlim(0, 255)

plt.subplot(1, 5, 5)
plt.hist(img4.ravel(), bins=256, range=(0, 255), color="purple", alpha=0.7)
plt.title("(a,b) = (0.5, 0)")
plt.xlim(0, 255)

plt.tight_layout()
plt.show()
```

```python
#Explanation:
# How a affects contrast with different values:
#     - a > 1: More contrast
#     - a < 1: Lower contrast
#     - a = 1: Unchange
# How b affects brightness with different values:
#     - b > 0: More brightness
#     - b < 0: Lower brightness
#     - b = 0: Unchange
```
