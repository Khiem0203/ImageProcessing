```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def info(img):
    print(f"Size: {img.shape[:2]}")
    channels = 1 if len(img.shape) == 2 else img.shape[2]
    print(f"Number of channels: {channels}")
    print(f"Data type: {img.dtype}")
    print(f"Min pixel: {np.min(img)}")
    print(f"Max pixel: {np.max(img)}")
```

```python
original_img = cv2.imread("./cat.jpg")
info(original_img)
og_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
#Plot origin image
plt.title("Original Image")
plt.axis("off")
plt.imshow(og_img)
```

```python
#Convert to grayscale
gray_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
cv2.imwrite("./cat_gray.jpg", gray_img)
info(gray_img)
#Plot grayscale image
plt.title("Grayscale Image")
plt.axis("off")
plt.imshow(gray_img, cmap="gray")
```

```python
#Resize image
height, width = gray_img.shape
resize_cal = (int(width*1.5), int(height*1.5))
resize_img = cv2.resize(gray_img, resize_cal, interpolation=cv2.INTER_LINEAR)
cv2.imwrite("./cat_upscale.jpg", resize_img)
info(resize_img)
#Plot resize grayscale image
plt.title("150% Grayscale Image")
plt.axis("off")
plt.imshow(resize_img, cmap="gray")
```

```python
#Histogram Grayscale
hitsto_gray = cv2.calcHist([gray_img], [0], None, [256], [0, 256])
plt.title("Histogram Grayscale Image")
plt.plot(hitsto_gray)
plt.show()
```

```python
#Equalized Image
equalized_img = cv2.equalizeHist(gray_img)
cv2.imwrite("./cat_gray_equalized.jpg", equalized_img)
#Plot equalized image
plt.title("Equalized Grayscale Image")
plt.imshow(equalized_img, cmap = "gray")
plt.axis("off")
```

```python
#Equalized Histogram
histo_equal = cv2.calcHist([equalized_img], [0], None, [256], [0, 256])
plt.title("Equalized Grayscale Image")
plt.plot(histo_equal)
plt.show()
```

```python
#Grayscale & Equalized Image side by side
plt.subplot(1, 2, 1)
plt.title("Grayscale")
plt.imshow(gray_img, cmap = "gray")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Equalized Grayscale")
plt.imshow(equalized_img, cmap = "gray")
plt.axis("off")
plt.show()
```
